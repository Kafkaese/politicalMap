```python
import pandas as pd
```

```python
df = pd.read_csv('../data/csv/countries.csv')
```

```python
df
```

```python
import geopandas
```

```python
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
```

```python
world
```

```python
df_comb = df.set_index('country name').join(world.set_index('name'), how = 'inner')
```

```python
world.set_index('name').merge(df.set_index('country name'), how='inner', left_index = True, right_index = True)
```

```python
df_comb.reset_index(names=['name'], inplace=True)
```

```python
df_comb.geometry
```

```python
import folium
```

```python
my_map = folium.Map()

# Add the data
folium.Choropleth(
    geo_data=df_comb,
    name='choropleth',
    data=df_comb,
    columns=['name', 'gdp_md_est'],
    key_on='feature.properties.name',
    fill_color='OrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name=''
).add_to(my_map)

my_map.save('meat.html')
```
